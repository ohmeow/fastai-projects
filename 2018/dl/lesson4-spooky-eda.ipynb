Text pre-processing steps:
1. **Tokenization** - Segregation of the txt into individual constituent words
2. **Stopwords** - Throw away any words that occurr too frequently as they are not useful in determining relevancy (also consider throwing away words that occur very infrequently)
3. **Stemming** - Combine variants of words into a single parent word that still conveys the same meaning
4. **Vectorization** - Converting text into a vector format.

### 2.3 Stemming and Lemmatization

Stemming attempts to reduce variations of similar words into a single term.  The words "running", "runs", and "run" would be simply referred to as "run".  *You do lose the granularity of the past, present, or future tenses)*.

NLTK provides various stemmers (Porter, Lancaster, Snowball)

```python
stemmer = nltk.stem.PorterStemmer()
print(stemmer.stem('leaves'))
```

Lemmatization attempts to do the same thing as stemming by reducing words based on an actual vocabulary (the Lemma) and will therefore not chop off words into stemmed forms that do not carry any lexical meaning.

```python
from nltk.stem import WordNetLemmatizer
lemm = WordNetLemmatizer()
print(lemm.lemmatize('leaves'))
```

## 3. Topic modeling vis LDA and NMF

```python
nlp = spacy.load('en')
from spacy.en.language_data import STOP_WORDS
```

When you vectorize the raw text with CountVectorizer, the dual stages of tokenizing and stopwords filtering are automatically included as a high-level component. It will:

1. Discard all single character terms like 'a', 'w', etc...
2. Conver text to lower case
3. Filter stopwords by setting "stop_words=english"

Because there is no built in lemmatizer, we have two options:

1. Lemmatize text everytime before feeding it into the vectorizer.
2. Extend CountVectorizer and overwrite "build_analyzer() to return lemmatized list (!!!)

```python
class LemmaCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(LemmaCountVectorizer, self).build_analyzer()
        return lambda doc: (nlp(w)[0].lemma_ for w in analyzer(doc))
```

```python
# store entire training text in list
corpus = list(train_df.text.values)
```

```python
# stop_words='english' : use sklearn's stopwords
# stop_words=STOP_WORDS : use spacy stopwords list
tf_vectorizer = LemmaCountVectorizer(max_df=0.95, min_df=2, stop_words=STOP_WORDS, decode_error='ignore')
tf = tf_vectorizer.fit_transform(corpus)
```

### 3.1 Latent Dirichlet Allocation

LDA is a unsupervised learning algorithm used for topic modeling.  It attempts to identify *K* topics given *D* documents and the *W* words in each document

```python
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(n_components=9, max_iter=5,
                                learning_method = 'online',
                                learning_offset = 50.,
                                random_state = 0)
```

```python
lda.fit(tf)
```

```python
# return the top 10 words attributed to each LDA generated topic
feature_names = tf_vectorizer.get_feature_names()
n_top_words = 50

for idx, topic in enumerate(lda.components_):
    words = " ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1 :-1]])
    print(words)
    print('-' * 100)
```

```python
first_topic = lda.components_[0]
second_topic = lda.components_[1]
third_topic = lda.components_[2]
fourth_topic = lda.components_[3]

first_topic.shape, second_topic.shape, third_topic.shape, fourth_topic.shape
```
